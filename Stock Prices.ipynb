```python
#Import libraries
import pandas as pd
import pandas_datareader as data
import datetime
from bokeh.plotting import figure, show, output_file
```

```python
start=datetime.datetime(2016,1,1) #start date of stock
end = datetime.datetime(2016,3,10) #end date of stocks
#examine google stock for the dates above
df = data.DataReader(name='GOOG',data_source='yahoo',start=start,end=end) #Create table with data pulled from Yahoo Finance
df.head(2)
```

```python
df.index[df.Close > df.Open] #index the dataframe to get dates only when the prices increased
```

```python
#Function to review stock performance
def incr_decr(c,o):
    if c>o:
        value='Increase'
    elif c<o:
        value = 'Decrease'
    else:
        value='Equal'
    return value
df['Performance']=[incr_decr(c,o) for c,o in zip(df.Close, df.Open)]
```

```python
df.head(4)
```

```python
#Create candlestick chart
p= figure(title='Google Stock Prices',x_axis_type='datetime',width=1000,height=300)
hours_12 = 12*60*60*1000

p.rect(df.index[df.Close > df.Open],(df.Open+df.Close)/2,hours_12,\
       abs(df.Open-df.Close), fill_color='green', line_color='black')

p.rect(df.index[df.Close < df.Open],(df.Open+df.Close)/2,hours_12,\
       abs(df.Open-df.Close),fill_color='green', line_color='black')

output_file('candlestick.html')
show(p)
```

```python
#Create candlestick chart
p= figure(title='Number of Increases and Decreases',x_axis_type='datetime',width=1000,height=300)
hours_12 = 12*60*60*1000

p.rect(df.index[df.Status== 'Increase'],(df.Open+df.Close)/2,hours_12,\
       abs(df.Open-df.Close), fill_color='green', line_color='black')

p.rect(df.index[df.Status=='Decrease'],(df.Open+df.Close)/2,hours_12,\
       abs(df.Open-df.Close),fill_color='green', line_color='black')

output_file('candlestick.html')
show(p)
```

```python
df['Daily Average']=(df.Open+df.Close)/2
df.head(4)
```

```python
df['Height']=abs(df.Close-df.Open)
df.head(2)
```

```python
#Create candlestick chart
name = ['Increase Dates','Decrease Dates']

p= figure(title='Change in Stock Price',x_axis_type='datetime',width=1000,height=500,sizing_mode='scale_width')
hours_12 = 12*60*60*1000

p.grid.grid_line_alpha=0.3

p.rect(df.index[df.Status== 'Increase'],(df.Open+df.Close)/2,hours_12,\
       abs(df.Open-df.Close), fill_color='#CCFFFF', line_color='black')

p.rect(df.index[df.Status=='Decrease'],(df.Open+df.Close)/2,hours_12,\
       abs(df.Open-df.Close),fill_color='#FF3333', line_color='black')

p.segment(df.index,df.High,df.index,df.Low,color='Black')

output_file('candlestick.html')
show(p)
```
